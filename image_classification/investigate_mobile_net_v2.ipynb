```python
import keras
from keras_applications import mobilenet_v2 
from keras.applications.imagenet_utils import decode_predictions
from keras.preprocessing import image
from keras.preprocessing.image import img_to_array, load_img

import numpy as np
import matplotlib.pyplot as plt
```

```python
model = MobileNetV2(weights='imagenet')
```

```python
#https://www.google.co.uk/search?q=couch&source=lnms&tbm=isch&sa=X&ved=0ahUKEwjulu7h7Y_cAhXCL8AKHa5MDJoQ_AUICygC&biw=1440&bih=723#imgrc=PlvxqaOXaM0FqM:
couch1_path = 'downloaded_images/couch1.jpeg'
```

```python
original = load_img(couch1_path, target_size=(224, 224))
```

```python
numpy_image = img_to_array(original)
plt.imshow(np.uint8(numpy_image))
plt.show()
```

```python
image_batch = np.expand_dims(numpy_image, axis=0)
processed_image = mobilenet_v2.preprocess_input(image_batch.copy())
```

```python
predictions = model.predict(processed_image)
```

```python
label = decode_predictions(preds=predictions, top=5)
```
